```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# URL of page to be scraped
imbd_url = 'https://www.imdb.com/search/title?groups=top_1000&count=250'
# Retrieve page with the requests module
response = requests.get(imbd_url)
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# # results are returned as an iterable list
imbdmovies = soup.find_all('h3', attrs={'class':'lister-item-header'})
print(imbdmovies)
```

```python
results = imbdmovies [0]
results
```

```python
movies = []
for movie in imbdmovies:
    title = movie.find('a').text
    year = movie.find('span', class_ = 'lister-item-year text-muted unbold').text[1:-1]

    movies.append((title, year))

moviesdf = pd.DataFrame(movies, columns=['Top IMBd Movies', 'Year of Release'])
moviesdf.head()
```

```python
moviesdf.to_csv('IMDbTopMovies.csv', index=False, encoding='utf-8')
```
