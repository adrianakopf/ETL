```python
from bs4 import BeautifulSoup
from splinter import Browser
import pandas as pd
import requests
```

```python
# URL of page to be scraped --> target info. Sorted by Popularity Ascending
url = 'https://www.imdb.com/search/title?groups=top_1000&count=250'

# Retrieve page with the requests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html.parser')

imbdmovies = soup.find_all('h3', attrs={'class':'lister-item-header'})
results = imbdmovies [0]
results
```

```python
#Chrome will "take charge" 
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path)
browser.visit(url)
```

```python
#Lists for dataframe
titles=[]
years=[]

#Scrape, print on screen, load info for pandas dataframe  
for x in range(1, 5):
 
    html = browser.html
    # Create BeautifulSoup object; parse with 'html.parser'
    soup = BeautifulSoup(html, 'html.parser')
    imbdmovies = soup.find_all('h3', attrs={'class':'lister-item-header'})

    for movie in imbdmovies:
        title = movie.find('a').text
        year = movie.find('span', class_ = 'lister-item-year text-muted unbold').text[1:-1]
        titles.append(title)
        years.append(year)
            
    #Print on screen
    for movie in imbdmovies:
        print(movie.text)
        
    #Go to next page to load/read more files
    browser.click_link_by_partial_text("Next")
   
```

```python

# create Pandas dataframe
moviesdf = pd.DataFrame({
    "title": titles,
    "year": years
})

# moviesdf.head()
moviesdf.tail()
```

```python
moviesdf.to_csv('IMDbTop1000Movies.csv', index=False, encoding='utf-8')
```
